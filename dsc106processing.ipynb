```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from collections import defaultdict 
```

```python
csvfile = pd.read_csv('./CompleteData.csv')
```

```python
csvfile = csvfile.fillna(0)
```

```python
csvfile
```

```python
csvfile['FL_DATE'] = pd.to_datetime(csvfile['FL_DATE'])
csvfile['FL_DATE_MONTH'] = csvfile['FL_DATE'].dt.month
```

```python
csvfile
```

```python
## get each airline and their average delay per month 
grouped_df = csvfile.groupby(['OP_UNIQUE_CARRIER','FL_DATE_MONTH']).mean()
```

```python
grouped_df.to_csv('monthly_data.csv')
```

```python
csvfile
```

```python
airports = np.unique(csvfile[['ORIGIN', 'DEST']].values.ravel())
```

```python
airports
```

```python
data = {'nodes': [], 'links': []}
for airport in airports:
    data['nodes'].append({'id': airport})
for _, row in csvfile.iterrows():
    origin = row['ORIGIN']
    dest = row['DEST']
    data['links'].append({'source': origin, 'target': dest})
```
